```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df['employment_type'] = df['employment_type'].map({
    'PT': 'Part-time',
    'FT': 'Full-time',
    'CT': 'Contract',
    'FL': 'Freelance'
})
df['experience_level'] = df['experience_level'].map({
    'EN': 'Junior',
    'MI': 'Intermediate',
    'SE': 'Expert',
    'EX': 'Director'
})

df['remote_ratio'] = df['remote_ratio'].map({
    0: 'WFO',
    50: 'Hybrid',
    100: 'WFH'
})
df['company_size'] = df['company_size'].map({
    'S': 'Startup',
    'M': 'Medium',
    'L': 'Big Enterprise'
})

df.head()
```

Which Role Has Highest Salary employment wise

```python
highest_salary=df.groupby('job_title').max()['salary_in_usd'].sort_values(ascending=False)
highest_salary
```

```python
top5=highest_salary[:5]
bottom5=highest_salary[-5:]
print(top5,'\n\n',bottom5)
```

Which country pay the highest for which roles?

```python
for role in top5.index:
  print(role,end=':')
  temp=df.loc[df['job_title']==role]['salary_in_usd'].idxmax()
  topcountry,sal=df.iloc[temp]['company_location'], df.iloc[temp]['salary_in_usd']
  print(topcountry,'$',sal)
  print()

```

YoY salary grouwth at different experience level

```python
data2020=df.loc[df['work_year']==2020]
data2021=df.loc[df['work_year']==2021]
data2022=df.loc[df['work_year']==2022]
data2023=df.loc[df['work_year']==2023]

for level in df.experience_level.unique():
    print(level)
    avg2020 = data2020.groupby('experience_level').mean().loc[level, 'salary_in_usd']
    avg2021 = data2021.groupby('experience_level').mean().loc[level, 'salary_in_usd']
    avg2022 = data2022.groupby('experience_level').mean().loc[level, 'salary_in_usd']
    avg2023 = data2023.groupby('experience_level').mean().loc[level, 'salary_in_usd']

    yoy20_21 = ((avg2021 - avg2020) / avg2020) * 100
    print('YoY Growth for 2020-2021: ', round(yoy20_21, 2))

    yoy21_22 = ((avg2022 - avg2021) / avg2021) * 100
    print('YoY Growth for 2021-2022: ', round(yoy21_22, 2))

    yoy22_23 = ((avg2023 - avg2022) / avg2022) * 100
    print('YoY Growth for 2021-2022: ', round(yoy22_23, 2))

    print()
```

```python
df["company"].value_counts()
```

```python
plt.title('job vacancy in companies')
plt.xlabel('comapny')
plt.ylabel('no of vacancy')
plt.barh(list(df["company"].value_counts().keys()[0:10]),list(df["company"].value_counts()[0:10]))
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('% age of top 10 comapanies having maximum job vacancy')
plt.pie(list(df["company"].value_counts()[0:10]),labels=list(df["company"].value_counts().keys()[0:10]),autopct="%0.2f")
plt.show()
```

```python
df["area"].value_counts()
```

```python
plt.title('job vacancy in areas')
plt.xlabel('areas')
plt.ylabel('no of vacancy')
plt.barh(list(df["area"].value_counts().keys()),list(df["area"].value_counts()))
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('%age of top areas by job')
plt.pie(list(df["area"].value_counts()),labels=list(df["area"].value_counts().keys()),autopct="%0.2f")
plt.show()
```
